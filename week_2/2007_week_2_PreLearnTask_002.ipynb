# 【問題1】乱数の作成

平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。


\begin{matrix} 1.0 & 0.8 \\ 0.8 & 1.0 \end{matrix}


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0) # シードの固定

mean = [-3, 0] # 平均を指定
cov = [[1.0, 0.8], [0.8, 1.0]] # 共分散行列の指定

ndarr = np.random.multivariate_normal(mean, cov, 500)
print(np.shape(ndarr))
```

# 【問題2】散布図による可視化

問題1で作成したデータ点を散布図により可視化してください。



```python
np.random.seed(0)
# 2次元正規乱数を500個生成
x, y = np.random.multivariate_normal(mean, cov, 500).T

plt.scatter(x, y)
plt.show()
```

# 【問題3】ヒストグラムによる可視化

# 【問題４】　データの追加
新たに平均が(0,-3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。


\begin{matrix} 1.0 & 0.8 \\ 0.8 & 1.0 \end{matrix}

そして問題1,4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1の物は0、問題2のものは1を表示してください。


```python
np.random.seed(0) # シードの固定

mean_2 = [0, -3] # 平均を指定
cov = [[1.0, 0.8], [0.8, 1.0]] # 共分散行列の指定

learn_four = np.random.multivariate_normal(mean_2, cov, 500)
x2, y2 = np.random.multivariate_normal(mean_2, cov, 500).T

plt.title("scatter")
plt.scatter(x, y, label="0")
plt.scatter(x2, y2, label="1")
plt.legend(bbox_to_anchor=(1,1))

plt.show()
```

# 【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを結合し、(1000, 2)のndarrayとしてください。

```python
union_arr = np.concatenate((ndarr, learn_four), axis=0)
print(np.shape(union_arr))
```

# 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。


0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。


機械学習に使用するデータセットはこのような形になっていることが多いです。



```python
tmp = np.insert(ndarr, 2, 0, axis=1)
tmp2 = np.insert(learn_four, 2, 1, axis=1)
label_arr = np.concatenate((tmp, tmp2), axis=0)
print(np.shape(label_arr))
print(label_arr)
```
