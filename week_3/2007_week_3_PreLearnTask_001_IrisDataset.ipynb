# 【問題1】データの取得
以下のドキュメントを参考にしながら、irisデータセット読み込みましょう。

[sklearn.datasets.load_iris — scikit-learn 0.20.2 documentation](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html)


データが読み込めたら、説明変数を下のようなPandasのDataFrame型変数Xに格納してください。列の名前も以下を参考にしてください。


[pandas.DataFrame — pandas 0.24.2 documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html)

```python
from sklearn.datasets import load_iris
import pandas as pd

RAW = load_iris()
#print(data)

X = pd.DataFrame(data=RAW["data"], columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
print(X)

y = pd.DataFrame(data=RAW["target"], columns=['Species'])
print(y)
```

# 【問題2】データの結合
問題1で準備したX,yを結合させたDataFrame型変数dfを作成してください。結合することで扱いやすくなります。これ以降はこのdfを使っていきます。

```python
df = X.join(y)
print(df)
```

### irisデータセットとは

“setosa”, “versicolor”, “virginica” という 3 種類の品種のアヤメのがく片 (Sepal)、花弁 (Petal) の幅および長さを計測したデータです。
150本の花のがく片、花弁の幅および長さをdata、各データの分類 (0: 'setosa', 1: 'versicolor', 2: 'virginica')をtargetとして保持しています。



>アヤメのデータセット、またはフィッシャーのアヤメデータセットは、英国の統計学者で生物学者ロナルド・フィッシャーが、1936年の論文で導入した分類学で利用する線形判別分析のデータセットです。 エドガー・アンダーソンは、3種のアヤメの形態変化を定量化するためデータ収集したため、アンダーソンのアヤメデータセットと呼ばれることがあります。 3種のうちの2種は、同じ日に、同じ牧草地、同じ装置で、同じ人が、同時に採取、測定したものです。

>データセットは、Iris（Iris setosa、Iris virginica、およびIris versicolor）の3種のそれぞれの50サンプルからなります。 各サンプルから4つの特徴である、がく（萼）と花弁（花びら）の長さと幅を測定しました。

>これらの4つの特徴の組み合わせに基づいて、フィッシャーは種を互いに区別するための線形判別モデルを開発しました。

[上記引用元](https://programmer-jobs.blogspot.com/2018/05/iris-data-set-used-in-machine-learning.html)


以上から、アヤメのデータセットとは、

<b>複数の特徴の組み合わせに基づいて、種類を区別するための線形判別モデルのためのデータセット</b>

# 【問題6】図の作成
データは可視化して確認することが大切です。次の3つを行ってください。作図のライブラリはMatplotlibの他seabornもよく使われます。


・ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）

・特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する

・特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

箱ひげ図とバイオリン図は似た図を表示します。これらはどう違い、それぞれどのような利点があるのかについて説明してください。

```python
# 円グラフ表示

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#matplotlib.pyplot.pie

grouped = df.groupby('Species')
species_prop = np.array(grouped.size())


plt.title("species_prop")
label = "setosa", "versicolor", "virginica" 
plt.pie(species_prop, labels=label, counterclock=False, startangle=90, autopct="%1.1f%%")
plt.show()
```

```python
# 箱ひげ図

fig, ax = plt.subplots()

columns = (df["sepal_length"], df["sepal_width"], df["petal_length"], df["petal_width"])
bp = ax.boxplot(columns)
plt.title("Iris Dataset")
ax.set_xticklabels(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])


plt.show()
```

```python
# バイオリン図

fig, ax = plt.subplots()

columns = (df["sepal_length"], df["sepal_width"], df["petal_length"], df["petal_width"])
bp = ax.violinplot(columns)
plt.title("Iris Dataset")
ax.set_xticklabels(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])


plt.show()
```

### 箱ひげ図とバイオリン図の違いについて

箱ひげ図は最大値と最小値がわかりやすく、バイオリン図は最大〜最小の間のデータ分布がわかりやすい。

箱ひげ図は外れ値が分かりやすいが、二峰性や一様分布といった特徴は一切排除されているため、各分布のデータが同一の特徴を持っているかのように見えてしまう。
バイオリン図は分布が分かりやすいがカーネル密度推定によって滑らかに表示されているため、実際は存在しないデータを錯覚する可能性がある。



# 【問題7】特徴量間の関係の確認
特徴量間の関係を見ていきます。次の4つを行ってください。


・縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る

・全ての組み合わせの散布図を一度に表示させる、散布図行列を作る

・４つの特徴量に関する相関係数行列を作る

・相関係数行列のヒートマップを作る

相関係数行列はDataFrameのcorr()メソッドを使うことで求められます。

```python
# 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
#sepal_width, sepal_length

setosa = df[df['Species'].isin([0])]
versicolor = df[df['Species'].isin([1])]
virginica = df[df['Species'].isin([2])]

plt.title("sepal")
plt.xlabel("width")
plt.ylabel("length")
x = setosa["sepal_width"]
y = setosa["sepal_length"]
x2 = versicolor["sepal_width"]
y2 = versicolor["sepal_length"]
x3 = virginica["sepal_width"]
y3 = virginica["sepal_length"]

plt.scatter(x, y, label="setora")
plt.scatter(x2, y2, label="versicolor")
plt.scatter(x3, y3, label="virginica")
plt.legend(bbox_to_anchor=(1.1, 1))

plt.show()
```

```python
#petal_width, petal_length
plt.title("petal")
plt.xlabel("width")
plt.ylabel("length")
x = setosa["petal_width"]
y = setosa["petal_length"]
x2 = versicolor["petal_width"]
y2 = versicolor["petal_length"]
x3 = virginica["petal_width"]
y3 = virginica["petal_length"]

plt.scatter(x, y, label="setora")
plt.scatter(x2, y2, label="versicolor")
plt.scatter(x3, y3, label="virginica")
plt.legend(bbox_to_anchor=(0.3, 1))

plt.show()
```

```python
#sepal_width, petal_width

plt.title("width")
plt.xlabel("sepal_width")
plt.ylabel("petal_width")
x = setosa["sepal_width"]
y = setosa["petal_width"]
x2 = versicolor["sepal_width"]
y2 = versicolor["petal_width"]
x3 = virginica["sepal_width"]
y3 = virginica["petal_width"]

plt.scatter(x, y, label="setora")
plt.scatter(x2, y2, label="versicolor")
plt.scatter(x3, y3, label="virginica")
plt.legend(bbox_to_anchor=(1.02, 1))

plt.show()
```

```python
#sepal_length, petal_length

plt.title("length")
plt.xlabel("sepal_length")
plt.ylabel("petal_length")
x = setosa["sepal_length"]
y = setosa["petal_length"]
x2 = versicolor["sepal_length"]
y2 = versicolor["petal_length"]
x3 = virginica["sepal_length"]
y3 = virginica["petal_length"]

plt.scatter(x, y, label="setora")
plt.scatter(x2, y2, label="versicolor")
plt.scatter(x3, y3, label="virginica")
plt.legend(bbox_to_anchor=(0.3, 1))

plt.show()
```

```python
# 散布図行列をプロット

import seaborn as sns

vdf = sns.load_dataset("iris")
sns.pairplot(df, hue="Species")
plt.show()
```

```python
# ４つの特徴量に関する相関係数行列を作る

corr_mat = df.corr(method='pearson')
print(corr_mat)
```

```python
sns.heatmap(corr_mat,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True,
            fmt='1.1f',
            xticklabels=corr_mat.columns.values,
            yticklabels=corr_mat.columns.values
           )
plt.show()
```
